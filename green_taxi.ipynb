```python
print('Maximum Trip distance',df['Trip_distance'].max())
print('Minimum Trip distance',df['Trip_distance'].min())
print('Median Trip distance',df['Trip_distance'].median())

z=np.asarray(df['Trip_distance'])
p=np.percentile(z,99.5)
print('The 99.5 percentile value for Trip Distance is',p)
```

```python
def outlier_iqr(feature, k=1.5):
    # calculate interquartile range 
    z = np.asarray(feature)
    q25, q75 = np.percentile(z, 25), np.percentile(z, 75)
    print('Quartile 25: {} | Quartile 75: {}'.format(q25, q75))
    iqr = q75 - q25
    print('iqr: {}'.format(iqr))

    # calculate the outlier cutoff bounds
    cut_off = iqr * k
    lower, upper = q25 - cut_off, q75 + cut_off
    print('Upper bound: {}'.format(upper))

    # exclude any data point located k standard deviations above the median point 
    outlier = [x for x in z if x > upper]
    print('Number of outliers for Trip Distance feature: {}'.format(len(outlier)))
    inlier = [x for x in z if x < upper]
    
    # Trip distance without outliers
    f, (ax1, ax2) = plt.subplots(1,2, figsize=(15, 5))
    
    sns.distplot(inlier, bins=30, color='b', hist_kws={"linewidth": 1, 'edgecolor':'black', "alpha": 0.7}, 
                      kde_kws={"linewidth": 2, "color":'black'}, ax=ax1)
    ax1.set_title('Distribution of Trip Distance (without outliers)')
    ax1.set_xlabel('Trip Distance (miles)')
    ax1.set_ylabel('Count')
    
    # Log of Trip distance without outliers
    # Remove zeros because log of 0 is undefined
    inlier_remove_zero = list(filter(lambda a: a != 0, inlier))
    sns.distplot(np.log(inlier_remove_zero), bins=30, color='b', hist_kws={"linewidth": 1, 'edgecolor':'black', "alpha": 0.7}, 
                      kde_kws={"linewidth": 2, "color":'black'}, ax=ax2)
    ax2.set_title('Log of Trip Distance (Without Outliers)')
    ax2.set_xlabel('Trip Distance (miles)')
    ax2.set_ylabel('Count')
  
outlier_iqr(df['Trip_distance'], k=3)
```



*   To the left is plotted the distribution of Trip distance with outliers removed.
*   To the right is a log of lognormal distribution seems to be a Gaussian distribution which also proves our hypothesis that the Trip distance is lognormally distributed.

We define the coordinates of the three major airports (JFK, LGA, EWR) and create a new feature "CheckAirport" to classify a taxi ride depending on whether the pick up or drop off location of a taxi ride is it to or from an airport.

```python
# Classify a taxi ride as JFK (John F Kennedy), LGA(LaGuardia), EWR (Newark Liberty) or NOT(Not an airport)
k = []

for index, row in df.iterrows():
    if  (
        ((row['Pickup_longitude'] < -73.76) & (row['Pickup_longitude'] > -73.785) &
         (row['Pickup_latitude'] < 40.65) & (row['Pickup_latitude'] > 40.63)) |
        ((row['Dropoff_longitude'] < -73.76) & (row['Dropoff_longitude'] > -73.785) &
         (row['Dropoff_latitude'] < 40.65) & (row['Dropoff_latitude'] > 40.63))
        ):
        k.append("JFK")
    elif  (
          ((row['Pickup_longitude'] < -73.87) & (row['Pickup_longitude'] > -73.88456) &
           (row['Pickup_latitude'] < 40.773) & (row['Pickup_latitude'] > 40.767)) |
          ((row['Dropoff_longitude'] < -73.87) & (row['Dropoff_longitude'] > -73.88456) &
           (row['Dropoff_latitude'] < 40.773) & (row['Dropoff_latitude'] > 40.767))
          ):
        k.append("LGA")
    elif  (
          ((row['Pickup_longitude'] < -74.13) & (row['Pickup_longitude'] > -74.19) &
           (row['Pickup_latitude'] < 40.72) & (row['Pickup_latitude'] > 40.67)) |
          ((row['Dropoff_longitude'] < -74.13) & (row['Dropoff_longitude'] > -74.19) &
           (row['Dropoff_latitude'] < 40.72) & (row['Dropoff_latitude'] > 40.67))
          ):
        k.append("EWR")
    else:
        k.append("NOT")

df['CheckAirport']=k
```

## Data Cleaning & Imputation

```python
df1 = df.copy()
```

```python
# Print unique values to check for invalid data 
print('Unique values in VendorID:', df1['VendorID'].unique())
print('Unique values in Store_and_fwd_flag:', df1['Store_and_fwd_flag'].unique())
print('Unique values in RateCodeID:', df1['RateCodeID'].unique())
print('Unique values in Passenger_count:', df1['Passenger_count'].unique())
print('Unique values in Extra:', df1['Extra'].unique())
print('Unique values in Payment_type:', df1['Payment_type'].unique())
print('Unique values in Trip_type:', df1['Trip_type '].unique())
```

```python
# Ehail_fee : Drop this feature since all of its values are NaNs
df1.drop('Ehail_fee', axis = 1, inplace = True)

# RateCodeID : Replace invalid data by most frequent value 1
df1.loc[~((df1.RateCodeID>=1) & (df1.RateCodeID<=6)), 'RateCodeID'] = 1

# Extra : Replace invalid data by most frequent value 0
df1.loc[~((df1.Extra==0) | (df1.Extra==0.5) | (df1.Extra==1)), 'Extra'] = 0

# Trip_type : Replace missing values with the most frequent value 1
df1['Trip_type '] = df1['Trip_type '].replace(np.NaN,1)

# Total_amount, Fare_amount, improvement_surcharge, Tip_amount, Tolls_amount, MTA_tax : Remove negative values 
df1.Total_amount = df1.Total_amount.abs()
df1.Fare_amount = df1.Fare_amount.abs()
df1.improvement_surcharge = df1.improvement_surcharge.abs()
df1.Tip_amount = df1.Tip_amount.abs()
df1.Tolls_amount = df1.Tolls_amount.abs()
df1.MTA_tax = df1.MTA_tax.abs()
```

There are a few rides where the Total amount is zero. In such cases, Tip amount as a percentage of Total amount tends to infinity. But we know that the minimum rate for a green cab is 2.5$. Thus any transaction with a Total amount less than 2.5 is replaced with the mean.

```python
# Replacing Total_amount < 2.5 with the mean (Data Imputation)
mean_Total_amount = df1.loc[df1['Total_amount'] >= 2.5, 'Total_amount'].mean()
df1['Total_amount'] = np.where(df1['Total_amount'] < 2.5, mean_Total_amount, df1['Total_amount'])

# Calculate Tip_percent
df1['Tip_percent'] = df1['Tip_amount']*100/df1['Total_amount']
print('Summary: Tip percentage\n', df1['Tip_percent'].describe())
```

## Feature Engineering

```python
# extract Pickup Week, Weekday, Monthday, Hour 
import datetime
ref_week = datetime.date(2015,9,1).isocalendar()[1]
df1['pickup_week'] = df1.lpep_pickup_datetime.apply(lambda x:x.isocalendar()[1]) - ref_week + 1
df1['pickup_weekday']  = df1.lpep_pickup_datetime.apply(lambda x:x.isocalendar()[2])
df1['pickup_monthday'] = df1.lpep_pickup_datetime.apply(lambda x:x.day)
df1['pickup_hour'] = df1.lpep_pickup_datetime.apply(lambda x:x.hour)

# exract Trip duration from pickup and dropoff timestamps
df1['Trip_duration'] = (df1.Lpep_dropoff_datetime-df1.lpep_pickup_datetime).apply(lambda x:x.total_seconds()/60)

# create Tip_Group variable
df1['Tip_Group'] = np.where(df1.Tip_percent>0, 1, 0)

# drop 'lpep_pickup_datetime' & 'Lpep_dropoff_datetime' since we extracted all the information in those columns
df1.drop(['lpep_pickup_datetime', 'Lpep_dropoff_datetime'], axis = 1, inplace = True)
```


*   Here I tried to convert the geographic latitudes and longitudes into more meaningful region labels using k-means clustering. Since k-means tries to group based solely on euclidean distance between objects, we will get back clusters of locations that are close to each other. To find the optimal number of clusters we make use of 'elbow' plot.



```python
loc_df = pd.DataFrame()
loc_df['longitude'] = longitude
loc_df['latitude'] = latitude
```

```python
# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 10):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42).fit(loc_df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 10), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```



*   The line chart looks like an arm and the "elbow" on the arm is the value of k that is the best. Here K=3

```python
# Fitting K-Means to the dataset
kmeans = KMeans(n_clusters=3, random_state=2, n_init = 10).fit(loc_df)
loc_df['label'] = kmeans.labels_

fig,ax = plt.subplots(figsize = (10,10))
for label in loc_df.label.unique():
    ax.plot(loc_df.longitude[loc_df.label == label],loc_df.latitude[loc_df.label == label],'.', alpha = 0.3, markersize = 0.3)
    ax.plot(kmeans.cluster_centers_[label,0],kmeans.cluster_centers_[label,1],'o', color = 'r')
    ax.annotate(label, (kmeans.cluster_centers_[label,0],kmeans.cluster_centers_[label,1]), color = 'b', fontsize = 20)
    
ax.set_title('Clusters of New York')
plt.show()
```

## Predictive modelling - Regression

*   Since there is no linear relationship between the the Tip percentage and variables, random forest algorithm will be considered to build the regression part of the model.
The main hyperparameters for random forests are:

> 1) The tree depth (max_depth)

> 2) The number of trees (n_estimators)

> 3) The minimum number of samples required to split a leaf (min_samples_split)

> 4) The minimum number of samples in a leaf (min_samples_leaf)

> 5) The maximum number of features for each tree (max_features)

*   Due to time constraints, only the first two hyperparameters will be tuned. The remaining parameters will be set to their defaults. We define a list of values for both max_depth and n_estimators and then iterate over each to build model for each possible combination. Here we iterate over small number of trees (5, 10, 30) but better results can be obtained with 100 or 150 trees. 

*   We Identify best hyperparameters by performing 5-fold cross validation on the training data. This allows us to better assess the model performance on unseen validation data. Once the best parameters are chosen, we predict on both train and test data. 

*   The first thing we should worry about is underfitting: training performance is bad. The next thing we should worry about is overfitting: training performance is good, but test performance is bad.

*   We use Mean square error as the performance metric for the Regression problem.


```python
df2 = df1.copy()
```

```python
# Tip_percent (Target Variable) was derived from Tip_amount and hence we no longer need Tip_amount. 
# We also drop Longitude and Latitude variables.
df2 = df2.drop(['Tip_amount', 'Tip_Group', 'Pickup_longitude', 'Pickup_latitude', 'Dropoff_longitude', 'Dropoff_latitude'], axis=1)
```

```python
# One-hot-encode categorical variables 
df2 = pd.get_dummies(df2, columns=qualitative)
```

```python
#separate dependent and independent variables
X = df2.drop('Tip_percent', axis=1)
y = df2['Tip_percent']
```

```python
# Split the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 12345)
```

```python
quantitative = ['Passenger_count', 'Trip_distance', 'Fare_amount', 'Extra', 'MTA_tax',
                'Tolls_amount', 'improvement_surcharge', 'Total_amount',
                'pickup_hour', 'pickup_weekday', 'pickup_week', 'pickup_monthday', 
                'Trip_duration', 'Trip_speed']

from sklearn.preprocessing import StandardScaler 
# Create standardizer
standardizer = StandardScaler()
# Fit standardizer to training set
standardizer.fit(X_train[quantitative])
# Apply to both training and test sets
X_train[quantitative] = standardizer.transform(X_train[quantitative])
X_test[quantitative] = standardizer.transform(X_test[quantitative])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_validate
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.metrics import median_absolute_error
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.metrics import f1_score, confusion_matrix, classification_report 
from sklearn.metrics import roc_curve, auc
import math

max_depth_list = [5, 10, 15]
n_trees_list = [5, 10, 30]
score_list = ['neg_mean_squared_error', 'neg_mean_absolute_error']
score_names = ['MSE', 'MAE']
min_mse = 1e64
for n_trees in n_trees_list:
  for d in max_depth_list:
    print("\nNumber of Trees: ", n_trees, " Max_Depth: ", d)
    rfr = RandomForestRegressor(n_estimators=n_trees, criterion='mse', max_depth=d, max_features='auto',
                                min_samples_split=2, n_jobs=1, random_state=12345)
    scores = cross_validate(rfr, X_train, y_train, scoring=score_list, return_train_score=False, cv=5)
    print("{:.<13s}{:>6s}{:>13s}".format("Metric", "Mean", "Std. Dev."))
    i=0
    for s in score_list:
      var = "test_"+s
      mean = math.fabs(scores[var].mean())
      std = scores[var].std()
      label = score_names[i]
      i += 1
      print ("{:.<13s}{:>7.4f}{:>10.4f}".format(label, mean, std))
      if label == 'MSE' and mean < min_mse:
        min_mse = mean
        best_depth = d
        best_n_trees = n_trees

print ("\nBest based on MSE from a forest with ", best_n_trees, " trees.")
print ("Best Depth (trees) = ", best_depth)


rfr = RandomForestRegressor(n_estimators=best_n_trees, criterion="mse", max_depth=best_depth, min_samples_split=2, 
                            min_samples_leaf=1, max_features='auto', n_jobs=1, bootstrap=True, random_state=12345)
rfr= rfr.fit(X_train, y_train)
```

```python
# Print evaluation metrics for both train and test data
def display_split_metrics(rfr, X_train, y_train, X_test, y_test):
  predict_train = rfr.predict(X_train)
  predict_test = rfr.predict(X_test)
  print("\n")
  print("{:.<23s}{:>15s}{:>15s}".format('Model Metrics', 'Training', 'Testing'))
  print("{:.<23s}{:15d}{:15d}".format('Observations', X_train.shape[0], X_test.shape[0]))
  print("{:.<23s}{:15d}{:15d}".format('Coefficients', X_train.shape[1]+1, X_test.shape[1]+1))

  R2train = r2_score(y_train, predict_train)
  R2test = r2_score(y_test, predict_test)
  print("{:.<23s}{:15.4f}{:15.4f}".format('R-Squared', R2train, R2test))
  print("{:.<23s}{:15.4f}{:15.4f}".format('Mean Absolute Error', mean_absolute_error(y_train,predict_train), mean_absolute_error(y_test,predict_test)))
  print("{:.<23s}{:15.4f}{:15.4f}".format('Median Absolute Error', median_absolute_error(y_train,predict_train), median_absolute_error(y_test,predict_test)))
  print("{:.<23s}{:15.4f}{:15.4f}".format('Avg Squared Error', mean_squared_error(y_train,predict_train), mean_squared_error(y_test,predict_test)))
  print("{:.<23s}{:15.4f}{:15.4f}".format('Square Root ASE', math.sqrt(mean_squared_error(y_train,predict_train)), math.sqrt(mean_squared_error(y_test,predict_test))))

display_split_metrics(rfr, X_train, y_train, X_test, y_test)
```

```python
predictors = X.columns.tolist()
feature_imp = pd.Series(rfr.feature_importances_,predictors).sort_values(ascending=False)
feature_imp.plot(kind='bar',title='Feature Importance')
plt.ylabel('Importance of the predictor')
plt.show()
```

**Results**

*   Mean square error on test sample: 8.1

*   As can be seen from the above Feature importance plot, Payment_type seems to play a major role and an importance feature to be slected along with Total_amount and Fare_amount.


## Predictive modelling - Classification

*   We bin the Tip percentage as per the following rules: 

>  Zero Tip - Class 0

> 0-19% Tip - Class 1

> Above 19% Tip - Class 2

*   We apply Logistic regression, Gradient Boosting classifier and Neural Networks to our classification problem to compare the results and identify the best algorithm for our problem. 

*   We start off with applying Logistic Regression. We tune the hyperparameters using Grid Search and simultaneously use 5-Fold cross validation. 

*   We then use Gradient Boosting classifier since this has been proven successful across many domains and is one of the leading methods for achieving better predictions. 

*  As we have a decently large dataset for our problem, Neural Networks can show their potential since they get better with more data we feed into them.

*  Both F1 score and ROC-AUC score can be used for class imbalance. We use F1 score  as the performance metric for the classification problem.

*  Our classifcation task at hand suffers from class-imbalance problem. Due to time constraints, we won't be focusing on applying resampling techniques. Oversampling technique such as SMOTE or a weighted loss function can help us overcome this problem. 



```python
# Divide tip percentage into classes
y_train = y_train.tolist()
y_test = y_test.tolist()

y_train = [0 if v == 0 else 1 if (v > 0) & (v <= 19) else 2 for v in y_train]
y_test = [0 if v == 0 else 1 if (v > 0) & (v <= 19) else 2 for v in y_test]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score, f1_score, roc_curve, roc_auc_score, accuracy_score, classification_report

# Logistic Regression 
log_reg_params = {"penalty": ['l2'], 'C': [0.1, 1, 10, 100]}
grid_log_reg = GridSearchCV(LogisticRegression(), log_reg_params, cv=5)
grid_log_reg.fit(X_train, y_train)
# Print the best parameters and estimator
print("Tuned Logistic Regression Parameters: {}".format(grid_log_reg.best_params_)) 
print("Tuned Logistic Regression Estimator: {}".format(grid_log_reg.best_estimator_)) 
```

```python
# Fit using the best estimator
optimised_log_reg = grid_log_reg.best_estimator_
optimised_log_reg = optimised_log_reg.fit(X_train, y_train)
predict_train = optimised_log_reg.predict(X_train)
predict_test = optimised_log_reg.predict(X_test)
```

```python
print("{:.<23s}{:>15s}{:>15s}".format('Model Metrics', 'Training', 'Testing'))
print("{:.<23s}{:15.4f}{:15.4f}".format('accuracy: ', accuracy_score(y_train,predict_train), accuracy_score(y_test,predict_test)))
print("{:.<23s}{:15.4f}{:15.4f}".format('precision: ', precision_score(y_train,predict_train, average='macro'), precision_score(y_test,predict_test, average='macro')))
print("{:.<23s}{:15.4f}{:15.4f}".format('recall: ', recall_score(y_train,predict_train, average='macro'), recall_score(y_test,predict_test, average='macro')))
print("{:.<23s}{:15.4f}{:15.4f}".format('f1: ', f1_score(y_train,predict_train, average='macro'), f1_score(y_test,predict_test, average='macro')))
```

```python
# Define Model 
from keras import models
from keras import layers
model = models.Sequential()
model.add(layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(3, activation='softmax'))

# Encoding the data
from keras.utils.np_utils import to_categorical
one_hot_y_train = to_categorical(y_train)
one_hot_y_test = to_categorical(y_test)

# Compiling the model
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])

# Training the model
history = model.fit(X_train, one_hot_y_train, epochs=20, batch_size=512, validation_split=0.3)
```

```python
## Need to check axes

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(16,6))

# Plotting the training and validation loss
import matplotlib.pyplot as plt
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(loss) + 1)
ax1.plot(epochs, loss, 'bo', label='Training loss')
ax1.plot(epochs, val_loss, 'b', label='Validation loss')
ax1.set_title('Training and validation loss')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss')
ax1.legend()

# Plotting the training and validation accuracy
acc = history.history['acc']
val_acc = history.history['val_acc']
ax2.plot(epochs, acc, 'bo', label='Training acc')
ax2.plot(epochs, val_acc, 'b', label='Validation acc')
ax2.set_title('Training and validation accuracy')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Loss')
ax2.legend()
plt.show()
```

We can observe that the training loss decreases with every epoch, and the training accuracy
increases with every epoch. That’s what we would expect when running gradient descent
optimization—the quantity we’re trying to minimize should be less with
every iteration. But that isn’t the case for the validation loss and accuracy: they seem to
diverge after the fourth epoch. A model that performs better on the training data isn’t necessarily a model that will do better on data it has never seen before. In precise terms, what we’re seeing is overfitting:
after the seventh epoch, we’re overoptimizing on the training data, and end
up learning representations that are specific to the training data and don’t generalize
to data outside of the training set. In this case, to prevent overfitting, we should stop training after seven epochs.

```python
#One-way ANOVA on weeks
from scipy import stats
weeks=df3['pickup_week'].unique()
f, p = stats.f_oneway(*(df3[df3['pickup_week'] == week]['Trip_speed'] for week in weeks))
f, p
```


*   The one-way ANOVA tests whether the mean of some numeric variable differs across the levels of one categorical variable. It essentially answers the question: do any of the group means differ from one another?   

*   The hypotheses of interest in an ANOVA are as follows:

H0: μ1 = μ2 = μ3 ... = μk
H1: Means are not all equal.

*   The test output yields an F-statistic of 2.47 and a p-value of 0.042, indicating that there is a significant difference between the means of each group. We can reject the null hypothesis and say that speeds are significantly different. 


*   To analyse further and find out which weeks differ in the average speed, post hoc test is performed.

*   Here the chosen post hoc test is the Tukey-HSD test, which compares 2 groups at a time.


```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.stats.multicomp import MultiComparison

mc = MultiComparison(df3.Trip_speed, df3.pickup_week)
result = mc.tukeyhsd()
print(result)
```

**Results:**

*   From the result table, we can see that we cannot reject the null hypothesis for any of the two weeks.

*   The above summary table suggests that the mean speed does not differ between different weeks of the month.
